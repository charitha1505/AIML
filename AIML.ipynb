```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

# Load the dataset
df = pd.read_csv('/housing (2).csv')
print(df.head())
df.info()

# Check for missing values
print(df.isnull().sum())

# Visualize relationship between Built_up_area and Rent
plt.scatter(df['Built_up_area'], df['Rent'], color='red', marker='*')
plt.xlabel('Built_up_area')
plt.ylabel('Rent')
plt.title('Scatter Plot: Built_up_area vs Rent')
plt.show()

# Prepare features and target
x = df.drop('Rent', axis=1)
y = df['Rent']

# Split into train and test sets
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.33, random_state=42)

print('xtrain shape:', xtrain.shape)
print('xtest shape:', xtest.shape)
print('ytrain shape:', ytrain.shape)
print('ytest shape:', ytest.shape)

# Train the model
reg = LinearRegression()
reg.fit(xtrain, ytrain)

# Make predictions
y_pred = reg.predict(xtest)


sample_input = pd.DataFrame([[650, 0, 1]], columns=x.columns)
predicted_rent = reg.predict(sample_input)
print(f"Predicted rent for input [650, 0, 1]: {predicted_rent[0]}")

# Model coefficients and intercept
print("Coefficients:", reg.coef_)
print("Intercept:", reg.intercept_)

# Evaluate the model
train_score = reg.score(xtrain, ytrain)
test_score = reg.score(xtest, ytest)
print('Train R-squared:', train_score)
print('Test R-squared:', test_score)

# Error metrics
mse = mean_squared_error(ytest, y_pred)
mae = mean_absolute_error(ytest, y_pred)
print('Mean Squared Error:', mse)
print('Mean Absolute Error:', mae)
```
